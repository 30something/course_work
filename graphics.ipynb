```python
import matplotlib.pyplot as plt
import numpy as np
X = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])

def plot(SOL, LB, G1, G2, G3, D, GR1, GR2, ymin, ymax, i, j, axs, m): 
    SOL = np.array(SOL)
    LB = np.array(LB)
    G1 = np.array(G1)
    G2 = np.array(G2)
    G3 = np.array(G3)
    D = np.array(D)
    
    axs[i, j].plot(X, SOL / LB)
    axs[i, j].scatter(X, SOL / LB, label = 'Near-optimal solution')

    axs[i, j].plot(X, G1 / LB)
    axs[i, j].scatter(X, G1 / LB, label = 'Greedy v1')

    axs[i, j].plot(X, G2 / LB)
    axs[i, j].scatter(X, G2 / LB, label = 'Greedy v2')

    axs[i, j].plot(X, G3 / LB)
    axs[i, j].scatter(X, G3 / LB, label = 'Greedy v3')
    
#     axs[i, j].plot(X, D / LB)
#     axs[i, j].scatter(X, D / LB, label = 'Dummy')

    axs[i, j].plot(X, GR1 / LB)
    axs[i, j].scatter(X, GR1 / LB, label = 'Two groups')
    
    axs[i, j].plot(X, GR2 / LB)
    axs[i, j].scatter(X, GR2 / LB, label = 'Three groups')

    axs[i, j].set_xlabel('n', fontsize = 14)
    axs[i, j].set_title(f'Result / LB (m = {m})', fontsize = 14, pad = 10)
    axs[i, j].set_xticks(np.linspace(100, 1000, 10))
    axs[i, j].tick_params(axis = 'both', which = 'major', labelsize=12)
    axs[i, j].set_ylim(ymin, ymax)
    axs[i, j].legend(fontsize = 14)
```
